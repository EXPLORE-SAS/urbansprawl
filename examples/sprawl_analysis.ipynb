### Get shapefiles for input cities

### Process sprawling indices

```python
# Spatial step for indices calculation in meters
step={'dispersion': 100, 'landusemix': 100, 'accessibility': 100}
# Process cities
for city_ref, file_poly, file_point in zip(city_refs, poly_files, point_files):
	process_spatial_indices(city_ref, file_poly, file_point, step=step)
```
